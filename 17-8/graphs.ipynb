```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

csv_data = pd.read_csv('./data.csv')
```

```python
aux: dict = {}
for item in csv_data["userId"].to_list():
    if item not in aux: aux[item] = 1
    else: aux[item] += 1

x = np.array([key for key in aux.keys()])
y = np.array([val for val in aux.values()])

plt.title(csv_data.columns[1])
plt.bar(x, y)
plt.show()
```
